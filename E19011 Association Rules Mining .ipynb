```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from apyori import apriori
```

```python
df = pd.read_csv(r'C:\Users\User\Desktop\Python Codes\lastfm.csv')
df.head()
```

### Association Rules Considering User and Artist alone

```python
# getting unique values of artists and user

artist=sorted(df.artist.unique())
user=sorted(df.user.unique())
```

```python
# Creating a transaction matrix

mat=pd.DataFrame(columns=artist)
mat["user"]=user

# storing values in transaction matrix
count =0

for i in user:
    
    temp=sorted(df[df["user"]==i]["artist"].unique())

    for j in temp:
        mat.loc[count,j]=j
    
    count=count+1
    
    print(i)
    
# storing the transaction matrix as csv
mat.to_csv("lastfm_transaction.csv")
```

```python
#Creating list of transaction lists as preprocessing for apriori algorithm
records=[]

for i in user:
    
    records.append(sorted(df[df["user"]==i]["artist"].unique()))

    print(i)
```

```python
help(apriori)
```

```python
association_rules = apriori(records, min_support=0.01, min_confidence=0.5, min_lift=3)
association_results = list(association_rules)
print(len(association_results))
association_results
```

```python
# Creating and writing the results into dataframe

count=0

disp=pd.DataFrame(columns=["Antecedent","Consequent","Support","Confidence","Lift"])

for result in association_results:
    
    disp.loc[count,]=[result[2][0][0],result[2][0][1],result[1],result[2][0][2],result[2][0][3]]
    count=count+1
```

```python
# viewing the dataframe
disp
```

```python
# sorting the rules in descending order of confidence

disp.sort_values(by=['Confidence'],ascending=False)
```

#### Female

```python
# extracting information of female users

df1=df[df.sex=="f"].copy()
df1.head()

artist1=sorted(df1.artist.unique())
user1=sorted(df1.user.unique())

#Creating list of transaction lists as preprocessing for apriori algorithm
records1=[]

for i in user1:
    
    records1.append(sorted(df1[df1["user"]==i]["artist"].unique()))
    
association_rules1 = apriori(records1, min_support=0.01, min_confidence=0.5, min_lift=3)
association_results1 = list(association_rules1)
print(len(association_results1))

# Creating and writing the results into dataframe

count=0

disp1=pd.DataFrame(columns=["Antecedent","Consequent","Support","Confidence","Lift"])

for result in association_results1:
    
    disp1.loc[count,]=[result[2][0][0],result[2][0][1],result[1],result[2][0][2],result[2][0][3]]
    count=count+1

# Sorting the rules in descending order of confidence
disp1.sort_values(by=['Confidence'],ascending=False)
```
